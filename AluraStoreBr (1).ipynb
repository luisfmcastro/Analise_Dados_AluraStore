### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
def faturamento(loja):
  faturamento = 0
  for venda in range(len(loja)):
    faturamento = faturamento + loja['Preço'][venda]
  return round(faturamento, 2)
```

```python
print(f'Faturamento da loja 1: {faturamento(loja)}')
print(f'Faturamento da loja 2: {faturamento(loja2)}')
print(f'Faturamento da loja 3: {faturamento(loja3)}')
print(f'Faturamento da loja 4: {faturamento(loja4)}')
```

# 2. Vendas por Categoria


```python
def vendas_por_categoria(loja):
  categorias = []
  for categoria in range(len(loja)):
    contador = 0
    for novacategoria in range(len(categorias)):
      if loja['Categoria do Produto'][categoria] == categorias[novacategoria]:
        contador += 1
    if contador == 0:
      categorias.append(loja['Categoria do Produto'][categoria])
  resultado = []
  for tipocategoria in range(len(categorias)):
    vezes = 0
    for incidencia in range(len(loja)):
      if loja['Categoria do Produto'][incidencia] == categorias[tipocategoria]:
        vezes += 1
    resultado.append((categorias[tipocategoria], vezes))
  return resultado
```

```python
print(f'Vendas por categoria da loja 1: {vendas_por_categoria(loja)}')
print(f'Vendas por categoria da loja 2: {vendas_por_categoria(loja2)}')
print(f'Vendas por categoria da loja 3: {vendas_por_categoria(loja3)}')
print(f'Vendas por categoria da loja 4: {vendas_por_categoria(loja4)}')

```

# 3. Média de Avaliação das Lojas

```python
def media_avaliacao(loja):
  soma = 0
  for avaliacao in range(len(loja)):
    soma = soma + loja['Avaliação da compra'][avaliacao]
  media = round(soma / len(loja), 2)
  return media
```

```python
print(f'Média de avaliação da loja 1: {media_avaliacao(loja)}')
print(f'Média de avaliação da loja 2: {media_avaliacao(loja2)}')
print(f'Média de avaliação da loja 3: {media_avaliacao(loja3)}')
print(f'Média de avaliação da loja 4: {media_avaliacao(loja4)}')
```

# 4. Produtos Mais e Menos Vendidos

```python
def contagem_produtos(loja):
  return loja['Produto'].value_counts()
```

```python
print(f'Contagem de vendas de produtos: ')
print(f'Loja 1: {contagem_produtos(loja)}')
print(f'Loja 2: {contagem_produtos(loja2)}')
print(f'Loja 3: {contagem_produtos(loja3)}')
print(f'Loja 4: {contagem_produtos(loja4)}')
```

# 5. Frete Médio por Loja


```python
def media_frete(loja):
  soma = 0
  for frete in range(len(loja)):
    soma = soma + loja['Frete'][frete]
  return round(soma / len(loja), 2)
```

```python
print(f'Média de frete da Loja 1: {media_frete(loja)}')
print(f'Média de frete da Loja 2: {media_frete(loja2)}')
print(f'Média de frete da Loja 3: {media_frete(loja3)}')
print(f'Média de frete da Loja 4: {media_frete(loja4)}')
```
